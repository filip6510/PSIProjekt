import bibliotek potrzebnych w projekcie


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import StratifiedKFold
```

Przetwarzanie danych.




```python
data = pd.read_csv('time_series_covid19_confirmed_global.csv');
print(data.head());
```

```python
europe = ["Ukraine","France","Spain","Sweden","Germany","Finland","Norway","Poland","Italy","United Kingdom","Romania","Belarus",
"Greece","Bulgaria","Iceland","Portugal","Czechia","Denmark","Hungary","Serbia","Austria","Ireland","Lithuania","Latvia","Croatia",
"Bosnia and Herzegovina","Slovakia","Estonia","Netherlands","Switzerland","Moldova","Belgium","Albania","North Macedonia","Slovenia",
"Montenegro","Cyprus","Luxembourg","Andorra","Malta","Liechtenstein","Guernsey","San Marino","Gibraltar","Monaco"]
```

Będziemy sprawdzać tylko państwa w Europie
Prowincje nas nie iteresują wiec usuwamy pierwsza kolumnę, nie mają dla nas też znaczenia współżędne geograficzne więc je też usuwamy


```python
data = data[data['Country/Region'].isin(europe)];
data.drop(data.columns[[0,2,3]], axis=1, inplace=True)
data.drop(data.iloc[:, 1:50], inplace = True, axis = 1)
print(data);
```

Sumujemy powtarzające się państwa aby dostać dane dotyczące całego kraju zamiast jego regionów


```python
data = data.groupby(['Country/Region']).sum()
#print(data)
```

Chcemy otrzymać dziene przypadki więc przekształcamy dane


```python
daily_cases = data.copy();
#print(len(daily_cases.columns) );
for i in range(len(daily_cases)):
  for x in range(len(daily_cases.columns) - 1, 1,-1):
    daily_cases.iloc[i,x] = daily_cases.iloc[i,x] - daily_cases.iloc[i,x -1];
print (daily_cases)
```

Model będzie uczony na danych z całej Europy poza Polską, ponieważ na danych z polski będziemy testować model

sprawdzamy czy nie ma ujemnych wartości i jeśli są ustawiamy je na 0

```python
daily_cases[daily_cases < 0] = 0
plt.figure(figsize=(12, 6), dpi=80)
plt.plot(daily_cases.sum())
plt.show()
```

```python
def create_dataframe(n):
  list = ['today cases'];
  for i in range (1,n,1):
    list.append('day-' +  i.__str__());
  return pd.DataFrame(columns=list);


def add_data(in_data, data_frame, n):
  for i in range(n,len(in_data),1):
    #print('step2 ' + i.__str__() + " " + n.__str__());
    list = [] 
    for j in range(i, i - n, -1):
      #print('step3 ' + i.__str__() + " " + j.__str__() );
      list.append(in_data.iloc[j]);
    data_frame.loc[len(data_frame)] = list  

```

Przygotowujemy dane do testu


```python

n_of_days = 15;
test = create_dataframe(n_of_days);
add_data(daily_cases.loc['Czechia'], test, n_of_days);
print (test);
```

przygotowujemy dane do trenowania

```python
train = create_dataframe(n_of_days);
aux = daily_cases.drop(['Czechia'])
for i in range(len(aux)):
  add_data(aux.iloc[i],train, n_of_days);
print (train)

```

```python
#from sklearn.model_selection import train_test_split
#X= df.loc[:, df.columns != 'today cases']
#Y= df['today cases'].astype(int);
#X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state = 3,test_size = 0.3)
```

```python
Y_train = train['today cases'].astype(float)
X_train = train.loc[:,train.columns != 'today cases'].astype(float);
Y_test = test['today cases'].astype(float)
X_test = test.loc[:,test.columns != 'today cases'].astype(float);
print (X_train);
print (Y_train);

print (Y_train.shape);
print (X_test);
print (Y_test);
print (Y_test.shape);
```

```python
seed=123
kfold = StratifiedKFold(n_splits=2, random_state=seed,shuffle=True)
grid_1 = GridSearchCV(make_pipeline(PolynomialFeatures(), Lasso(random_state=seed)),
                    param_grid={'polynomialfeatures__degree': [1, 2, 3, 4],
                    'lasso__alpha': [0.01, 0.1, 1, 10]
                               },
                    cv=kfold,
                    refit=True)
grid_1.fit(X_train, Y_train)
grid_1.best_params_
```

Sprawdzamy dokładność


```python
print(X_train)
print(Y_train)
print(X_test)
print(Y_test)
```

```python
from sklearn import linear_model
from sklearn.metrics import r2_score
lasso = linear_model.Lasso(alpha=10)#grid_1.best_params_['lasso__alpha'])
lasso.fit(X_train, Y_train)
print("traing :" + lasso.score(X_train, Y_train).__str__())
print("test : " + lasso.score(X_test, Y_test).__str__())
#print("test : " + r2_score(lasso.predict(X_test), Y_test).__str__())
plt.figure(figsize=(12, 6), dpi=80)
plt.plot(lasso.predict(X_test), label= 'prediction')
plt.plot(Y_test, label = 'actual')
plt.legend(title='Parameter where:')
plt.figure()

plt.show()
```

```python
from sklearn.linear_model import Ridge

rid = Ridge(alpha=10)
rid.fit(X_train, Y_train)
print("traing :" + rid.score(X_train, Y_train).__str__())
print("test : " + rid.score(X_test, Y_test).__str__())
plt.figure(figsize=(12, 6), dpi=80)
plt.plot(rid.predict(X_test), label= 'prediction')
plt.plot(Y_test, label = 'actual')
plt.legend(title='Parameter where:')
plt.figure()

plt.show()
```

Jak widać Ridge oraz lasso dają podbne rezultaty, sprawdźmy teraz dla modeli z sieciami neutonowymi



```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout,Flatten, LSTM
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers


model = Sequential();
model.add(Dense(64, input_shape = (X_train.shape[1],),activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
#model.add(Dense(64, activation='relu'))
#model.add(Dense(64, activation='relu'))

model.add(Dense(1))
model.compile(loss='mse', optimizer='adam',metrics=['mean_absolute_percentage_error'])
model.summary()
```

```python
model.fit(X_train, 
          Y_train,
          epochs=50,
          )
```

```python
#print(model.predict(X_test).tolist())

plt.figure(figsize=(12, 6), dpi=80)
plt.plot(model.predict(X_test), label= 'prediction')
plt.plot(Y_test, label = 'actual')
plt.legend(title='Parameter where:')
plt.figure()

plt.show()
print ("accuracy for test: " + r2_score(model.predict(X_test), Y_test).__str__())
```
